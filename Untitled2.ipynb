```python
def two_sum(nums, target):
    vistos = {}

    for i, num in enumerate(nums):
        falta = target - num

        if falta in vistos:
            return (vistos[falta], i)

        vistos[num] = i



print(two_sum([2, 7, 4], 6))
```
